```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#load datasets
df1 = pd.read_csv('1- share-healthy-diet-unaffordable.csv')
df2 = pd.read_csv('2- food-expenditure-per-person-per-year.csv')
df3 = pd.read_csv('3- hare-of-consumer-expenditure-spent-on-food.csv')
df4 = pd.read_csv('4- share-calorie-diet-unaffordable.csv')

print(df1.head())
print(df2.head())
print(df3.head())
print(df4.head())
```

```python
#check datasets
df = [df1, df2, df3, df4]
for i, df in enumerate (df, start=1):
    print(f"df{i}'s information:")
    print(df.info())
    print(f"df{i} summary statistics:")
    print(df.describe())
    print("\n")
```

```python
df = [df1, df2, df3, df4]
#mising values
for i, df in enumerate (df, start=1):
    print(f"df{i}'s number of missing values:")
    print(df.isnull().sum())

#missing values are code of region
```

```python
#trend analysis over the years
df2['Year'] = pd.to_datetime(df2['Year'], format='%Y').dt.year
df2_avg = df2.groupby("Year")["Total food expenditure"].mean().reset_index()
df2_avg.head()
plt.plot(df2_avg["Year"],df2_avg["Total food expenditure"], label="Line", color="blue", linestyle='-', marker='o')
plt.xticks(df2['Year'])
plt.xlabel("Year")
plt.ylabel("Average food expenditure")
plt.title("Food expenditure over the years")
plt.show()
```

```python
#Regional Analysis - Top 10

#price trends between different reions over the years
df2_entity = df2.groupby(["Entity","Year"])["Total food expenditure"].mean().reset_index()
top_10_df2 = df2_entity.groupby("Entity")["Total food expenditure"].mean().nlargest(10).index
df2_top10 = df2_entity[df2_entity["Entity"].isin(top_10_df2)]
df2_top10_pivot = df2_top10.pivot_table(values="Total food expenditure", index="Entity", columns="Year", aggfunc="mean")
plt.figure(figsize=(10,6))
sns.heatmap(df2_top10_pivot, annot=False, fmt=".2f", cmap="YlGnBu")
plt.xlabel('Year')
plt.ylabel('Region')
plt.title('Average Food Expenditure by Region and Year')
plt.show()

#compare affordability levels across different regions
df1_entity = df1.groupby("Entity")["Share of the population who cannot afford a healthy diet"].mean().reset_index()
df1_entity = df1_entity.sort_values("Share of the population who cannot afford a healthy diet", ascending=False).head(10)
plt.figure(figsize=(16,8))
sns.barplot(x='Entity', y='Share of the population who cannot afford a healthy diet', data=df1_entity, palette='muted')
plt.xlabel('Region')
plt.ylabel('Average share of population cannot afford a healthy diet')
plt.title('Average share of population cannot afford by regions')
plt.show()
```

```python
df1_mean = df1.groupby(['Entity'])['Share of the population who cannot afford a healthy diet'].mean().reset_index()
df4_mean = df4.groupby(['Entity'])['Share of the population who cannot afford sufficient calories'].mean().reset_index()
merged_df = pd.merge(df1_mean, df3, on ='Entity')
merged_df = pd.merge(merged_df, df4_mean, on = 'Entity')
merged_df = merged_df.rename(columns={'Share of the population who cannot afford a healthy diet':'Healthy Diet Afford', 'Share of the population who cannot afford sufficient calories':'Calories Afford', 'Food as share of total consumer expenditure':'Food Expenditure'})

import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Filter top countries based on Calories Afford
top_countries_ca = merged_df.nlargest(20, 'Calories Afford')['Entity']

# Filter the main dataframe to include only the top countries
filtered_df = merged_df[merged_df['Entity'].isin(top_countries_ca)]
filtered_df = filtered_df.drop_duplicates(subset=['Entity', 'Year'])

# Create a list of unique years
years = filtered_df['Year'].unique()

fig, ax = plt.subplots(figsize=(10, 6))

def update(frame):
    ax.clear()
    year_data = filtered_df[filtered_df['Year'] == years[frame]]
    ax.plot(year_data['Entity'], year_data['Food Expenditure'], marker='o', label='Food Expenditure')
    ax.plot(year_data['Entity'], year_data['Healthy Diet Afford'], marker='o', label='Healthy Diet Afford')
    ax.plot(year_data['Entity'], year_data['Calories Afford'], marker='o', label='Calories Afford')
    ax.set_xlabel('Entity')
    ax.set_ylabel('Value')
    ax.set_title(f'Year: {years[frame]}')
    ax.legend()

# Create the animation with the number of frames equal to the number of unique years
ani = FuncAnimation(fig, update, frames=len(years), repeat=True)

# Display the animation using IPython.display
HTML(ani.to_jshtml())
```
